# Simulación del Truco

Las **cartas** deben registra el numero, palo y valor de la carta. 
El valor debe reflejar la jerarquía de las cartas en el juego del truco.

```python
def calcular_valor(numero, palo):
    valores = {
        (1, 'espada'): 14,
        (1, 'basto'): 13,
        (7, 'espada'): 12,
        (7, 'oro'): 11,
        (3, 'espada'): 10, (3, 'basto'): 10, (3, 'oro'): 10, (3, 'copa'): 10,
        (2, 'espada'): 9, (2, 'basto'): 9, (2, 'oro'): 9, (2, 'copa'): 9,
        (1, 'oro'): 8, (1, 'copa'): 8,
        (12, 'espada'): 7, (12, 'basto'): 7, (12, 'oro'): 7, (12, 'copa'): 7,
        (11, 'espada'): 6, (11, 'basto'): 6, (11, 'oro'): 6, (11, 'copa'): 6,
        (10, 'espada'): 5, (10, 'basto'): 5, (10, 'oro'): 5, (10, 'copa'): 5,
        (7, 'basto'): 4, (7, 'copa'): 4,
        (6, 'espada'): 3, (6, 'basto'): 3, (6, 'oro'): 3, (6, 'copa'): 3,
        (5, 'espada'): 2, (5, 'basto'): 2, (5, 'oro'): 2, (5, 'copa'): 2,
        (4, 'espada'): 1, (4, 'basto'): 1, (4, 'oro'): 1, (4, 'copa'): 1,
    }
    return valores.get((numero, palo), 0)
 
class Carta:
    def __init__(self, numero, palo):
          self.numero = numero
          self.palo = palo
          self.valor = calcular_valor(numero, palo)

    def __str__(self):
        return f"Carta={self.numero} de {self.palo} (valor={self.valor})"
        

c = Carta(1, "oro")
print(c)
```

**Mazo**: Representan las cartas españolas de 1 al 12, en los palos de oro, copa, espada y basto.

Debe existir una función que permita barajar las cartas.
y otra función que permita repartir las cartas a los jugadores.

```python
import random
numeros = [1, 2, 3, 4, 5, 6, 7, 10, 11, 12]  
palos = ["espada", "oro", "basto", "copa"]

class Carta:
    def __init__(self, numero, palo):
        self.numero = numero
        self.palo = palo
        
    def __str__(self):
        return f"{self.numero} de {self.palo}"

class Mazo:
    def __init__(self):
        self.cartas = [Carta(numero, palo) for numero in numeros for palo in palos]
        
    def mezclar(self):
         random.shuffle(self.cartas)
        
    def repartir(self, cantidad):
         return [self.cartas.pop() for _ in range(cantidad)]


m = Mazo()
m.mezclar()
cartas_repartidas = m.repartir(3)

for carta in cartas_repartidas:
    print(carta)
```

**Jugador**: Deben tener un nombre y las cartas que tiene en la mano. Debe acumular los puntos ganados.

Nota: Para simplificar el problema, el jugador elige la carta a jugar de forma aleatoria.

```python
class Jugador:
    
    def __init__(self, nombre):
         self.nombre = nombre
         self.cartas = []
         self.puntos = 0
        
    def recibir_cartas(self, cartas):
        self.cartas = cartas
    
    def jugar_carta(self):
         if not self.cartas:
            return None
         carta_jugada = random.choice(self.cartas)
         self.cartas.remove(carta_jugada)
         return carta_jugada
    def sumar_puntos(self, puntos):
        self.puntos += puntos

    def __str__(self):  
        return f"Jugador: {self.nombre}\nPuntos: {self.puntos}\nCartas: {', '.join(map(str, self.cartas))}"
        
m = Mazo()
m.mezclar()


j = Jugador("Juan")
j.recibir_cartas(m.repartir(3))

print(f"Carta jugada: {j.jugar_carta()}")
print(j)
```

**Partida**: Debe tener dos jugadores, un mazo de cartas y un puntaje. Debe permitir jugar una carta por ronda y definir el ganador de la ronda. Debe permitir jugar hasta que se terminen las cartas del mazo.

```python
class Partida:
    def __init__(self, jugador1, jugador2):
         self.jugador1 = jugador1
         self.jugador2 = jugador2
         self.mazo = Mazo()  
         self.mazo.mezclar()  
         self.puntos = {jugador1.nombre: 0, jugador2.nombre: 0}  
        

    def jugar_ronda(self):
         carta_j1 = self.jugador1.jugar_carta()  
         carta_j2 = self.jugador2.jugar_carta() 
         if carta_j1 is None or carta_j2 is None:  
            print("Uno de los jugadores se ha quedado sin cartas.")
            return False  
        
         print(f"{self.jugador1.nombre} juega {carta_j1}")  
         print(f"{self.jugador2.nombre} juega {carta_j2}")  

        
         if carta_j1.numero > carta_j2.numero:  
            print(f"{self.jugador1.nombre} gana la ronda")
            self.puntos[self.jugador1.nombre] += 1 
         elif carta_j2.numero > carta_j1.numero:  
            print(f"{self.jugador2.nombre} gana la ronda")
            self.puntos[self.jugador2.nombre] += 1  
         else:
            print("Empate en la ronda")  
        
         return True  

        

    def jugar(self):
        
        while len(self.mazo.cartas) >= 6 and (self.puntos[self.jugador1.nombre] < 15 and self.puntos[self.jugador2.nombre] < 15):
            
            self.jugador1.recibir_cartas(self.mazo.repartir(3))
            self.jugador2.recibir_cartas(self.mazo.repartir(3))

            for _ in range(3):
                if not self.jugar_ronda():  
                    break
            print(f"Puntaje: {self.jugador1.nombre} {self.puntos[self.jugador1.nombre]} - {self.jugador2.nombre} {self.puntos[self.jugador2.nombre]}")

    
        if self.puntos[self.jugador1.nombre] >= 15:
            print(f"{self.jugador1.nombre} ha ganado la partida!")
        elif self.puntos[self.jugador2.nombre] >= 15:
            print(f"{self.jugador2.nombre} ha ganado la partida!")
        else:
            print("No hay más cartas, la partida ha terminado sin un ganador claro.")
       
p = Partida(Jugador("Mica"), Jugador("Gonza"))
p.jugar()
```
