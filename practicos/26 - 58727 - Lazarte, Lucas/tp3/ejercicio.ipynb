## Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- fecha:    10 lugares
- producto: 30 lugares
- precio:   10 lugares
- cantidad:  5 lugares

hacer una funcion que carge los en una lista de diccionarios. 

nota: con 
```python
lineas = open("datos.dat", "r").readlines() 
```
puede obtener la lista de string de todos los datos 

````python
def cargar_datos():
    lista_datos = []
    lineas = open("datos.dat","r").readlines()
    
    for linea in lineas:
        fecha = linea[0:10].strip()          
        producto = linea[10:40].strip()      
        precio = float(linea[40:50].strip()) 
        cantidad = int(linea[50:55].strip())  
        
        diccionario = {
            "fecha": fecha,
            "producto": producto,
            "precio": precio,
            "cantidad": cantidad
        }
        
        lista_datos.append(diccionario)
    
    return lista_datos
datos = cargar_datos()

print(datos)
````

## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def calcular_totales(datos):
    total_importe = 0.0
    total_cantidad = 0
    
    for registro in datos:
        importe_vendido = registro["precio"] * registro["cantidad"]
        
        total_importe += importe_vendido
        total_cantidad += registro["cantidad"]
    
    return total_importe, total_cantidad

importe, cantidad = calcular_totales(datos)

print(f"Las ventas fueron de ${importe:.2f} en {cantidad} unidades")
````

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def unidades_vendidas(datos):
    unidades_por_producto = {}
    
    for registro in datos:
        producto = registro["producto"]
        cantidad = registro["cantidad"]
        
        if producto in unidades_por_producto:
            unidades_por_producto[producto] += cantidad
        else:
            unidades_por_producto[producto] = cantidad
    
    return unidades_por_producto

def listar_ventas(unidades):
    print("Unidades vendidas por producto:")
    for producto, cantidad in unidades.items():
        print(f"{producto}: {cantidad} unidades")

datos = cargar_datos()

ventas = unidades_vendidas(datos)
listar_ventas(ventas)
````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def ranking_productos(datos, top=3):
    unidades_por_producto = {}

    for registro in datos:
        producto = registro["producto"]
        cantidad = registro["cantidad"]

        if producto in unidades_por_producto:
            unidades_por_producto[producto] += cantidad
        else:
            unidades_por_producto[producto] = cantidad

    ranking = sorted(unidades_por_producto.items(), key=lambda item: item[1], reverse=True)

    return ranking[:top]

def listar_ranking(ranking):
    print(f"Ranking de los {len(ranking)} productos mas vendidos:")
    for producto, cantidad in ranking:
        print(f"{producto}: {cantidad} unidades vendidas")

ranking = ranking_productos(datos)
listar_ranking(ranking)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
def ventas_por_mes(datos):
    ventas_mensuales = {}


    for registro in datos:
        fecha = registro["fecha"]
        producto = registro["producto"]
        cantidad = registro["cantidad"]

        # formato 'YYYY-MM-DD'
        mes_anio = f"{fecha[0:7]}"  
        if mes_anio not in ventas_mensuales:
            ventas_mensuales[mes_anio] = {}

        if producto in ventas_mensuales[mes_anio]:
            ventas_mensuales[mes_anio][producto] += cantidad
        else:
            ventas_mensuales[mes_anio][producto] = cantidad

    return ventas_mensuales
def listar_ventas_mensuales(ventas):
    print("Ventas por mes:")
    for mes, productos in sorted(ventas.items()):
        print(f"\nMes: {mes}")
        for producto, cantidad in productos.items():
            print(f"{producto}: {cantidad} unidades vendidas")

ventas = ventas_por_mes(datos)
listar_ventas_mensuales(ventas)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
def resumen_ventas(datos):
    total_precios = {}
    total_cantidades = {}
    total_importes = {}

    for registro in datos:
        producto = registro["producto"]
        precio = registro["precio"]
        cantidad = registro["cantidad"]
        importe_vendido = precio * cantidad

        if producto in total_precios:
            total_precios[producto] += precio * cantidad
            total_cantidades[producto] += cantidad
            total_importes[producto] += importe_vendido
        else:
            total_precios[producto] = precio * cantidad
            total_cantidades[producto] = cantidad
            total_importes[producto] = importe_vendido

    resumen = {
        producto: {
            "precio_promedio": total_precios[producto] / total_cantidades[producto],
            "cantidad_vendida": total_cantidades[producto],
            "importe_total": total_importes[producto]
        }
        for producto in total_precios
    }

    return resumen

def informe_ventas(resumen):
    print("Informe de Ventas:")
    for producto in sorted(resumen.keys()):
        datos_producto = resumen[producto]
        print(f"{producto}:")
        print(f"  Precio Promedio: ${datos_producto['precio_promedio']:.2f}")
        print(f"  Cantidad Vendida: {datos_producto['cantidad_vendida']} unidades")
        print(f"  Importe Total: ${datos_producto['importe_total']:.2f}")

resumen = resumen_ventas(datos)
informe_ventas(resumen)
````
