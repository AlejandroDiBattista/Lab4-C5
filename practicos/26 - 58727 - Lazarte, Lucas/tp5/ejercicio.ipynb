## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad):
        self._codigo = None
        self._nombre = None
        self._precio = None
        self._tipo = None
        self._cantidad = None

        # Asigna los valores utilizando los setters para validar
        self.codigo = codigo
        self.nombre = nombre
        self.precio = precio
        self.tipo = tipo
        self.cantidad = cantidad

    # Propiedad para 'codigo'
    @property
    def codigo(self):
        return self._codigo

    @codigo.setter
    def codigo(self, valor):
        if isinstance(valor, str) and valor.isdigit() and len(valor) == 4:
            self._codigo = valor
        else:
            raise ValueError("El código tiene que ser una cadena de 4 dígitos.")

    # Propiedad para 'nombre'
    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, valor):
        if isinstance(valor, str) and 1 <= len(valor) <= 100:
            self._nombre = valor
        else:
            raise ValueError("El nombre tiene tener entre 1 y 100 caracteres.")

    # Propiedad para 'precio'
    @property
    def precio(self):
        return self._precio

    @precio.setter
    def precio(self, valor):
        if isinstance(valor, (int, float)) and 10 <= valor <= 10000:
            self._precio = valor
        else:
            # No actualiza el precio si el valor es inválido
            pass

    # Propiedad para 'tipo'
    @property
    def tipo(self):
        return self._tipo

    @tipo.setter
    def tipo(self, valor):
        if isinstance(valor, str) and len(valor) <= 20:
            self._tipo = valor
        else:
            raise ValueError("El tipo debe ser una cadena de hasta 20 caracteres.")

    # Propiedad para 'cantidad'
    @property
    def cantidad(self):
        return self._cantidad

    @cantidad.setter
    def cantidad(self, valor):
        if isinstance(valor, int) and 0 <= valor <= 100:
            self._cantidad = valor
        else:
            # No actualiza la cantidad si el valor es inválido
            pass

    # Método para calcular el valor total
    def valorTotal(self):
        return self.precio * self.cantidad
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self, descripcion='', codigos=None, tipos=None):
        self.descripcion = descripcion
        self.codigos = codigos if codigos is not None else []
        self.tipos = tipos if tipos is not None else []

    def esAplicable(self, producto, cantidad):
        """Determina si la oferta es aplicable al producto dado."""
        if self.codigos and producto.codigo in self.codigos:
            return True
        if self.tipos and producto.tipo in self.tipos:
            return True
        return False

    def calcularDescuento(self, producto, cantidad):
        """Calcula el descuento aplicado al producto. Debe ser implementado por las subclases."""
        raise NotImplementedError("Este método debe ser implementado por la subclase.")


# Clase OfertaDescuento que hereda de Oferta
class OfertaDescuento(Oferta):
    def __init__(self, descuento, descripcion='Descuento 10%', codigos=None, tipos=None):
        super().__init__(descripcion=descripcion, codigos=codigos, tipos=tipos)
        self.descuento = descuento  # Descuento porcentual

    def calcularDescuento(self, producto, cantidad):
        """Calcula el descuento porcentual aplicado al producto."""
        if self.esAplicable(producto, cantidad):
            return producto.precio * cantidad * (self.descuento / 100)
        return 0


# Clase Oferta2x1 que hereda de Oferta
class Oferta2x1(Oferta):
    def __init__(self, descripcion='Oferta 2x1', codigos=None, tipos=None):
        super().__init__(descripcion=descripcion, codigos=codigos, tipos=tipos)

    def calcularDescuento(self, producto, cantidad):
        """Calcula el descuento aplicado en una oferta 2x1."""
        if self.esAplicable(producto, cantidad):
            # Número de productos que se obtienen gratis
            productos_gratis = cantidad // 2
            return productos_gratis * producto.precio
        return 0
```

```python
## NO MODIFICAR ESTE CODIGO ##

p1 = Producto('1234', 'Coca Cola', 1000, 'gaseosa', 10)
p2 = Producto('1235', 'Oreo',      2300, 'galleta', 10)

o10d = OfertaDescuento(10, codigos=['1234'])
assert o10d.calcularDescuento(p1, 10) == 1000 
assert o10d.calcularDescuento(p1, 1) == 100

assert o10d.calcularDescuento(p2, 10) == 0

o2x1 = Oferta2x1(tipos=['galleta'])
assert o2x1.calcularDescuento(p1, 10) == 0

assert o2x1.calcularDescuento(p2, 1) == 0
assert o2x1.calcularDescuento(p2, 2) == 2300
assert o2x1.calcularDescuento(p2, 3) == 2300
assert o2x1.calcularDescuento(p2, 4) == 4600
assert o2x1.calcularDescuento(p2, 5) == 4600

print("Prueba pasada exitosamente!")
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
class Catalogo:
    def __init__(self):
        self.productos = {}  # Diccionario para acceso rápido por código
        self.ofertas = []    # Lista de ofertas en orden de registro

    # Método para leer el catálogo desde un archivo CSV
    @classmethod
    def leer(cls, archivo):
        catalogo = cls()
        try:
            with open(archivo, newline='', encoding='utf-8') as csvfile:
                reader = csv.DictReader(csvfile)
                for row in reader:
                    producto = Producto(
                        codigo=row['codigo'],
                        nombre=row['nombre'],
                        precio=float(row['precio']),
                        tipo=row['tipo'],
                        cantidad=int(row['cantidad'])
                    )
                    catalogo.agregar(producto)
        except FileNotFoundError:
            print(f"Archivo {archivo} no encontrado. Se creará un catálogo vacío.")
        return catalogo

    # Método para guardar el catálogo en un archivo CSV
    def guardar(self, archivo):
        with open(archivo, 'w', newline='', encoding='utf-8') as csvfile:
            fieldnames = ['codigo', 'nombre', 'precio', 'tipo', 'cantidad']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            writer.writeheader()
            for producto in self.productos.values():
                writer.writerow({
                    'codigo': producto.codigo,
                    'nombre': producto.nombre,
                    'precio': producto.precio,
                    'tipo': producto.tipo,
                    'cantidad': producto.cantidad
                })

    # Método para agregar un producto al catálogo
    def agregar(self, *productos):
        for producto in productos:
            if producto.codigo in self.productos:
                raise ValueError(f"Producto con código {producto.codigo} ya existe en el catálogo.")
        self.productos[producto.codigo] = producto


    # Método para buscar un producto por código
    def buscar(self, codigo):
        return self.productos.get(codigo, None)

    # Método para registrar una oferta
    def registrarOferta(self, oferta):
        self.ofertas.append(oferta)

    # Método para buscar una oferta aplicable al producto y cantidad
    def buscarOferta(self, producto, cantidad):
        for oferta in self.ofertas:
            if oferta.esAplicable(producto, cantidad):
                return oferta
        return None

    # Método para calcular el descuento de una oferta aplicable
    def calcularDescuento(self, producto, cantidad):
        oferta = self.buscarOferta(producto, cantidad)
        if oferta:
            return oferta.calcularDescuento(producto, cantidad)
        return 0

    # Propiedad para la cantidad de productos en el catálogo
    @property
    def cantidadProductos(self):
        return len(self.productos)

    # Propiedad para la cantidad total de unidades en el catálogo
    @property
    def cantidadUnidades(self):
        return sum(producto.cantidad for producto in self.productos.values())

    # Propiedad para el valor total del catálogo sin descuentos
    @property
    def valorTotal(self):
        return sum(producto.valorTotal() for producto in self.productos.values())

    # Método para generar el informe del catálogo
    def informe(self):
        total_productos = self.cantidadProductos
        total_unidades = self.cantidadUnidades
        precio_promedio = self.valorTotal / total_unidades if total_unidades else 0
        valor_total = self.valorTotal

        # Agrupar por tipo
        tipos_dict = defaultdict(list)
        for producto in self.productos.values():
            tipos_dict[producto.tipo].append(producto)

        tipos_informe = ""
        for tipo, productos in tipos_dict.items():
            unidades = sum(p.cantidad for p in productos)
            precio_promedio_tipo = sum(p.precio for p in productos) / len(productos) if productos else 0
            tipos_informe += f"  - {tipo:<20}: {unidades}u x $ {precio_promedio_tipo:.2f}\n"

        # Listar ofertas
        ofertas_informe = ""
        for oferta in self.ofertas:
            ofertas_informe += f"  - {oferta.descripcion}\n"

        informe = (
            "INFORME CATALOGO\n"
            f"Cantidad de productos:   {total_productos}\n"
            f"Cantidad de unidades:    {total_unidades}\n"
            f"Precio Promedio:       $ {precio_promedio:.2f}\n"
            f"Valor total:           $ {valor_total:.2f}\n"
            "Tipos de productos: \n" +
            tipos_informe +
            "Ofertas:\n" + ofertas_informe
        )
        return informe

    # Método para vender un producto (reduce la cantidad en existencia)
    def vender(self, producto, cantidad):
        if not isinstance(producto, Producto):
            raise ValueError("El producto debe ser una instancia de la clase Producto.")
        if cantidad <= 0:
            raise ValueError("La cantidad a vender debe ser positiva.")
        if producto.cantidad < cantidad:
            raise ValueError(f"No hay suficiente stock para vender {cantidad} unidades del producto {producto.codigo}.")
        producto.cantidad -= cantidad

    def __repr__(self):
        return f"Catalogo({list(self.productos.values())})"
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba del catálogo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidades == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe
assert catalogo.cantidadUnidades == 97
assert catalogo.valorTotal == 163400

# Buscar por código
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidades == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
import re
class Cliente:
    def __init__(self, nombre, cuit):
        self._nombre = None
        self._cuit = None
        self.nombre = nombre  # Utiliza el setter para validar
        self.cuit = cuit      # Utiliza el setter para validar

    @property
    def nombre(self):
        return self._nombre

    @nombre.setter
    def nombre(self, valor):
        if isinstance(valor, str) and 1 <= len(valor) <= 100:
            self._nombre = valor
        # Si el valor no es válido, no se realiza el cambio (se ignora)

    @property
    def cuit(self):
        return self._cuit

    @cuit.setter
    def cuit(self, valor):
        if isinstance(valor, str) and self._validar_cuit(valor):
            self._cuit = valor
        # Si el valor no es válido, no se realiza el cambio (se ignora)

    def _validar_cuit(self, cuit):
        """
        Valida que el CUIT tenga el formato XX-XXXXXXXX-X,
        donde X son dígitos.
        """
        patron = r'^\d{2}-\d{8}-\d$'
        return bool(re.match(patron, cuit))

    def __repr__(self):
        return f"Cliente(nombre='{self.nombre}', cuit='{self.cuit}')"
```

```python
class Factura:
    # Variable de clase para mantener el último número de factura
    ultimo_numero = 0

    @classmethod
    def ultimaFactura(cls, n):
        """
        Establece el último número de factura.
        """
        cls.ultimo_numero = n

    @classmethod
    def nuevoNumero(cls):
        """
        Incrementa y retorna el siguiente número de factura.
        """
        cls.ultimo_numero += 1
        return cls.ultimo_numero

    def __init__(self, catalogo, cliente):
        """
        Inicializa una nueva factura con un número secuencial, fecha actual,
        cliente y catálogo de productos.
        """
        self.numero = Factura.nuevoNumero()
        self.fecha = datetime.now().strftime('%d/%m/%Y')
        self.cliente = cliente
        self.catalogo = catalogo
        self.productos = {}  # Diccionario para almacenar productos y cantidades

    def agregar(self, producto, cantidad):
        """
        Agrega una cantidad de un producto a la factura.
        Si el producto ya está en la factura, incrementa la cantidad.
        También reduce el stock en el catálogo.
        """
        # Primero, verificar y reducir el stock en el catálogo
        self.catalogo.vender(producto, cantidad)

        # Luego, agregar al diccionario de productos en la factura
        if producto in self.productos:
            self.productos[producto] += cantidad
        else:
            self.productos[producto] = cantidad

    @property
    def cantidadProductos(self):
        """
        Retorna la cantidad de productos únicos en la factura.
        """
        return len(self.productos)

    @property
    def cantidadUnidades(self):
        """
        Retorna la cantidad total de unidades vendidas en la factura.
        """
        return sum(self.productos.values())

    @property
    def subtotal(self):
        """
        Calcula el subtotal de la factura sin descuentos.
        """
        return sum(producto.precio * cantidad for producto, cantidad in self.productos.items())

    @property
    def descuentos(self):
        """
        Calcula el total de descuentos aplicados en la factura.
        """
        return sum(self.catalogo.calcularDescuento(producto, cantidad) for producto, cantidad in self.productos.items())

    @property
    def total(self):
        """
        Calcula el total de la factura después de aplicar descuentos.
        """
        return self.subtotal - self.descuentos

    def imprimir(self):
        """
        Genera un texto formateado para imprimir la factura.
        """
        lines = []
        lines.append(f"Factura: {self.numero}")
        lines.append(f"Fecha  : {self.fecha}")
        lines.append(f"Cliente: {self.cliente.nombre} ({self.cliente.cuit})\n")

        # Detalles de productos
        for producto, cantidad in self.productos.items():
            precio = producto.precio
            subtotal_producto = precio * cantidad
            lines.append(f"- {cantidad} {producto.nombre:<25} x ${precio} = ${subtotal_producto}")

            # Obtener la oferta aplicable
            oferta = self.catalogo.buscarOferta(producto, cantidad)
            if oferta:
                descuento = self.catalogo.calcularDescuento(producto, cantidad)
                lines.append(f"      {oferta.descripcion:<35} - ${descuento:.2f}")

        # Resumen de la factura
        lines.append("\n" + f"{'Subtotal:':>50} ${self.subtotal:.2f}")
        lines.append(f"{'Descuentos:':>50} ${self.descuentos:.2f}")
        lines.append(f"{'-'*55}")
        lines.append(f"{'TOTAL:':>50} ${self.total:.2f}")

        return "\n".join(lines)

    def __repr__(self):
        return (f"Factura(numero={self.numero}, fecha='{self.fecha}', cliente={self.cliente}, "
                f"cantidadProductos={self.cantidadProductos}, cantidadUnidades={self.cantidadUnidades}, "
                f"subtotal={self.subtotal}, descuentos={self.descuentos}, total={self.total})")
```
