## Productos
- `codigo`: 4 digitos
- `nombre`: 1 a 100 caracteres
- `precio`: 10 a 10000
- `tipo`: 0 a 20 caracteres
- `cantidad`: 0 a 100 
- `valorTotal`: cantidad * precio

```python
class Producto:
    def __init__(self, codigo, nombre, precio, tipo, cantidad): 
        self.codigo = codigo
        self.nombre = nombre
        self.Precio = precio
        self.tipo = tipo
        self.Cantidad = cantidad
    
    @property
    def valorTotal(self):
        return self.Precio * self.Cantidad
    @property
    def precio(self):
        return self.Precio
    @precio.setter
    def precio(self, valor):
        if valor<0:
            print("Valor negativo inválido")
        else:
            self.Precio = valor
    @property
    def cantidad(self):
        return self.Cantidad
    @cantidad.setter
    def cantidad(self, valor):
        if valor<0:
            print("Valor negativo inválido")
        else:
            self.Cantidad = valor
```

```python
## NO MODIFIQUE ESTE CODIGO ##

# Prueba de la clase Producto

p1 = Producto('0001', 'Coca Cola', 1500, 'gaseosa', 10)

assert p1.codigo == '0001'
assert p1.nombre == 'Coca Cola'
assert p1.precio == 1500

# Calcula el valor total 
assert p1.valorTotal == 15000 

# Asegura que los valores de precio y cantidad sean validos
p1.precio = -200
assert p1.precio == 1500    # Rechaza el valor negativo

p1.precio = 2000 
assert p1.precio == 2000

p1.cantidad = -5
assert p1.cantidad == 10    # Rechaza el valor negativo

p1.cantidad = 50
assert p1.cantidad == 50
assert p1.valorTotal == 100000

print("Prueba pasada exitosamente!")
```

# Ofertas
Debe permitir aplicar oferctas a codigos espeficicos de productos y a tipos de productos
- `descripcion`: 1 a 100 caracteres
- `codigos`: lista de codigos de productos
- `tipos`: lista de tipo de producto
- `esAplicable(producto, cantidad)`: retorna si la oferta es aplicable a ese producto
- `aplicar(producto, cantidad)`: retorna el precio final del producto con la oferta aplicada


```python
class Oferta:
    def __init__(self,descripcion, codigos = [], tipos = []): 
        self.descripcion = descripcion 
        self.codigos = codigos
        self.tipos = tipos 

    def calcularDescuento(self, producto, cantidad):    
        return 0

class OfertaDescuento(Oferta):
    def __init__(self, descuento, *args, **kwargs):
        self.descripcion = kwargs.get('descripcion', '')
        self.codigos = kwargs.get('codigos', [])
        self.tipos = kwargs.get('tipos', [])
        self.descuento = descuento
        super().__init__(self.descripcion,*args, **kwargs)
        

    def calcularDescuento(self, producto, cantidad):
        if producto.codigo in self.codigos or producto.tipo in self.tipos:
            total = producto.precio * cantidad
            return total * (self.descuento / 100)
        return 0
    
class Oferta2x1(Oferta):
    def __init__(self, *args, **kwargs):
        self.descripcion = kwargs.get('descripcion', '')
        self.codigos = kwargs.get('codigos', [])
        self.tipos = kwargs.get('tipos', [])
        super().__init__(self.descripcion,*args, **kwargs)
       

    def calcularDescuento(self, producto, cantidad):
        if producto.tipo in self.tipos or producto.codigo in self.codigos:
            if cantidad > 1:
                cantidad_producto = cantidad // 2 + cantidad % 2 
                return producto.precio * (cantidad - cantidad_producto)
        return 0


```

```python
## NO MODIFICAR ESTE CODIGO ##

p1 = Producto('1234', 'Coca Cola', 1000, 'gaseosa', 10)
p2 = Producto('1235', 'Oreo',      2300, 'galleta', 10)

o10d = OfertaDescuento(10, codigos=['1234'])
assert o10d.calcularDescuento(p1, 10) == 1000 
assert o10d.calcularDescuento(p1, 1) == 100

assert o10d.calcularDescuento(p2, 10) == 0

o2x1 = Oferta2x1(tipos=['galleta'])
assert o2x1.calcularDescuento(p1, 10) == 0

assert o2x1.calcularDescuento(p2, 1) == 0
assert o2x1.calcularDescuento(p2, 2) == 2300
assert o2x1.calcularDescuento(p2, 3) == 2300
assert o2x1.calcularDescuento(p2, 4) == 4600
assert o2x1.calcularDescuento(p2, 5) == 4600

print("Prueba pasada exitosamente!")
```

# Catalogo
- `leer(archivo) `    : Carga los productos desde el archivo
- `guardar(archivo)`  : Guarda los productos en el archivo
- `agregar(producto)` : Agrega un producto al catalogo
- `buscar(codigo)`    : Busca un producto por codigo o None si no existe
- `registrarOferta(oferta)`  : Registra una oferta
- `buscarOferta(producto, cantidad)`: Busca una oferta por codigo o None si no existe
- `calcularDescuento(producto, cantidad)`: Calcula el descuento de una oferta
- `cantidadProductos`: Retorna la cantidad de productos en el catalogo
- `cantidadUnidades`: Retorna la cantidad de unidades en el catalogo
- `valorTotal`: retorna el valor total del catalogo sin descuentos
- `informe()`: retorna un string con el informe del catalogo 

```python
import csv

class Catalogo:
    def __init__(self): 
        self.cantidadUnidadesCatalogo=0
        self.valorTotal=0
        self.catalogo=[]
        self.ofertas=[]

    @property
    def cantidadUnidades(self):
        return self.cantidadUnidadesCatalogo
        
    @property
    def cantidadProductos(self):
        return len(self.catalogo)
    @property
    def precio_promedio(self):
        return self.valorTotal / self.cantidadUnidadesCatalogo
    
    def tipo_productos(self):
        tipos = []
        for producto in self.catalogo:
            if producto.tipo not in tipos:
                 tipos.append(producto.tipo)
        return tipos
    def actualizar_Unidades(self):
        total_unidades = sum(producto.cantidad for producto in self.catalogo)  
        self.cantidadUnidadesCatalogo = total_unidades

    def actualizarValorTotal(self):
        total_valor = sum(producto.precio * producto.cantidad for producto in self.catalogo)
        self.valorTotal = total_valor

    def agregar(self, *producto):
        self.catalogo.extend(producto)
        self.actualizar_Unidades()
        self.actualizarValorTotal()

    def contabilizar_productos(self,nombre):
        cantidad = []
        for productos in self.catalogo:
            if productos.tipo == nombre:
                cantidad.append(productos.nombre)
        return cantidad

    def registrarOferta(self,oferta):
        self.ofertas.append(oferta)
    def calcularDescuentoOferta2x1(self, productos):
     total_descuento = 0
     for producto in productos:
        if isinstance(producto, Producto):
            for oferta in self.ofertas:
                if isinstance(oferta, Oferta2x1):
                    total_descuento += oferta.calcularDescuento(producto, producto.cantidad)
     return total_descuento
    def calcularDescuento(self, producto, cantidad):
     descuentos = []
     for oferta in self.ofertas:
        descuento = oferta.calcularDescuento(producto, cantidad)
        if descuento > 0:
            descuentos.append(descuento)
     return max(descuentos) if descuentos else 0

    def guardar(self, nombre):
        with open(nombre, mode='w', newline='') as archivo_csv:
            escritor = csv.writer(archivo_csv)
            escritor.writerow(['codigo', 'nombre', 'precio', 'tipo', 'cantidad'])
            for producto in self.catalogo:
                escritor.writerow([producto.codigo, producto.nombre, producto.precio, producto.tipo, producto.cantidad])
    def vender(self,producto, cantidad):
        for product in self.catalogo:
            if product.codigo == producto.codigo:
                product.cantidad -= cantidad
                self.actualizar_Unidades()
                self.actualizarValorTotal()
                
        

    def informe(self):  
        return (
        f"Cantidad de productos: {self.cantidadProductos}\n"
        f"Cantidad de unidades: {self.cantidadUnidadesCatalogo}\n"
        f"Precio Promedio: {self.precio_promedio}\n"
        f"Valor total: {self.valorTotal}\n"
        f"Tipos de productos: {self.tipo_productos()}\n"
        f"Gaseosa: {self.contabilizar_productos('gaseosa')}\n"
        f"Galleta: {self.contabilizar_productos('galleta')}\n"
        f"Ofertas: {[oferta.codigos for oferta in self.ofertas]}\n"
        f"Oferta 2x1: {[oferta.tipos for oferta in self.ofertas if isinstance(oferta, Oferta2x1)]}\n"
    )

    def buscar(self,codigo):
        for producto in self.catalogo:
            if producto.codigo == codigo:
                return producto
        return None
    @classmethod
    def leer(cls,nombre):
        catalogoclase = cls()
        with open(nombre, mode='r') as catalogo:
            productos = csv.reader(catalogo)
            next(productos) 
            for producto in productos:
                codigo, nombre, precio, tipo, cantidad = producto
                nuevo_producto = Producto(codigo, nombre, float(precio), tipo, int(cantidad))
                catalogoclase.agregar(nuevo_producto)
        return catalogoclase

    
```

```python

## NO MODIFIQUE ESTE CODIGO ##

# Prueba del catálogo 

catalogo = Catalogo()
p1 = Producto('0001', 'Coca Cola',  1500, 'gaseosa', 10)
p2 = Producto('0002', 'Pepsi Cola', 1200, 'gaseosa', 20)
p3 = Producto('0003', 'Sonrisa',    1200, 'galleta', 30)
p4 = Producto('0004', 'Oreo',       2300, 'galleta', 40)

## Agregar productos al catalogo 
catalogo.agregar(p1)
catalogo.agregar(p2)
catalogo.agregar(p3)
catalogo.agregar(p4)

assert catalogo.cantidadProductos == 4
assert catalogo.cantidadUnidadesCatalogo == 100

assert catalogo.valorTotal == 167000

## Calcular descuentos segun las ofertas registradas
assert catalogo.calcularDescuento(p1, 5) == 0
assert catalogo.calcularDescuento(p2, 5) == 0

# Ofertas no acumulables 
catalogo.registrarOferta(Oferta2x1(tipos=['galleta']))
catalogo.registrarOferta(OfertaDescuento(10, codigos=['0001', '0003']))

assert catalogo.calcularDescuento(p1, 5) == 750
assert catalogo.calcularDescuento(p2, 5) == 0
assert catalogo.calcularDescuento(p3, 5) == 2400

assert catalogo.valorTotal == 167000.0
catalogo.guardar('catalogo-prueba.csv') ## Guardar datos antes de vender

# Vender afecta la cantidad de unidades y el valor total
catalogo.vender(p3, 3)   

# Verificar que el informe se genere correctamente

informe = catalogo.informe()
assert "Cantidad de productos: " in informe
assert "Cantidad de unidades: " in informe
assert "Precio Promedio: " in informe
assert "Valor total: " in informe
assert "Tipos de productos: " in informe
assert "gaseosa" in informe
assert "galleta" in informe
assert "Ofertas:" in informe 
assert "Oferta 2x1" in informe
assert catalogo.cantidadUnidadesCatalogo == 97
assert catalogo.valorTotal == 163400

# Buscar por código
assert catalogo.buscar('0001') == p1
assert catalogo.buscar('0002') == p2
assert catalogo.buscar('0099') is None 

# Recuperar los datos guardados  
c2 = Catalogo.leer('catalogo-prueba.csv')

assert c2.cantidadProductos == 4
assert c2.cantidadUnidadesCatalogo == 100

# Valor antes de guardar
assert c2.valorTotal == 167000.0

print("Prueba pasada exitosamente!")
```

# Cliente
- `nombre`: 1 a 100 caracteres
- `cuit`: 13 digitos (formato XX-XXXXXXXX-X)

```python
class Cliente: 
    def __init__(self, nombre, cuit):
        self.Nombre = nombre
        self.Cuit = cuit 
    @property
    def nombre(self):
        return self.Nombre
    @nombre.setter
    def nombre(self, valor):
        if valor == '':
            print("No se puede dejar el nombre vacío")
        else:
            self.Nombre = valor
    @property
    def cuit(self):
        return self.Cuit
    @cuit.setter
    def cuit(self, valor):
        formato = "-1234567890"
        if formato in valor:
            self.Cuit = valor
        else:
            print("Formato incorrecto")
```

```python
## NO MODIFICAR ESTE CODIGO ##

# Prueba de la clase Cliente #

c1 = Cliente('Juan Perez', '20-12345678-1')

assert c1.nombre == 'Juan Perez'
assert c1.cuit   == '20-12345678-1'

c1.nombre = ''
assert c1.nombre == 'Juan Perez' # Rechaza el valor vacio

c1.nombre = 'Juana Perez'        # Acepta el nuevo valor
assert c1.nombre == 'Juana Perez'

c1.cuit = '1234567890123'
assert c1.cuit == '20-12345678-1' # Rechaza el valor incorrecto

c1.cuit = 'CC-12345678-1'
assert c1.cuit == '20-12345678-1' # Rechaza el valor incorrecto

print("Prueba pasada exitosamente!")
```

```python
class Factura:
    numero = 0
    def __init__(self, catalogo, cliente): 
        self.catalogo = catalogo
        self.cliente = cliente
        self.productos = []
        self.Numero = self.nuevoNumero()
        self.cantidadUnidades=0
        self.subtotal=0
        self.total=0
        self.descuentos=0
        self.total=0

    @property
    def numero(self):
        return self.Numero
    def agregar(self,productos, cantidad):
         existe = False
         for p in self.productos:
            if p.codigo == productos.codigo:
                p.cantidad += cantidad
                existe = True
                break
         if not existe:
            producto_copia = Producto(productos.codigo, productos.nombre, productos.precio, productos.tipo,productos.cantidad)
            producto_copia.cantidad = cantidad  
            self.productos.append(producto_copia)
         self.actualizarUnidades()
         self.aplicar_subtotal()
         self.aplicar_descuentos()
         self.aplicar_total()
         if existe:
            self.catalogo.vender(productos, cantidad)
         else:
             self.catalogo.vender(productos, cantidad)
    @classmethod
    def ultimaFactura(cls,numero):
        cls.numero = numero
    @classmethod
    def nuevoNumero(cls):
        cls.numero +=1
        return cls.numero
    @property
    def cantidadProductos(self):
        return len(self.productos)
    def actualizarUnidades(self):
        self.cantidadUnidades = sum(producto.cantidad for producto in self.productos)

    def aplicar_subtotal(self):
        total = sum(p.precio * p.cantidad for p in self.productos)
        self.subtotal = total
    def aplicar_descuentos(self):
         total_descuentos = 0
         for producto in self.productos:
            descuento = self.catalogo.calcularDescuento(producto, producto.cantidad)
            total_descuentos += descuento 
         self.descuentos = total_descuentos
    def aplicar_total(self):
        self.total= self.subtotal - self.descuentos 
   
    def imprimir(self):
     nombre_cantidad = ', '.join(f"{producto.cantidad} {producto.nombre}" for producto in self.productos)
     descuento_2x1 = self.catalogo.calcularDescuentoOferta2x1(self.productos)
     oferta_2x1_str = f"Oferta 2x1: {descuento_2x1:.2f}" 
     descuentos_str = f"Descuento 10%: {self.descuentos:.2f}" 
     return (
        f"Cliente: {self.cliente.nombre}\n"
        f"Productos: {nombre_cantidad}\n"
        f"{descuentos_str}\n"
        f"{oferta_2x1_str}\n"
        f"Subtotal: {self.subtotal:.2f}\n"
        f"TOTAL: {self.total:.2f}"
     )

```
