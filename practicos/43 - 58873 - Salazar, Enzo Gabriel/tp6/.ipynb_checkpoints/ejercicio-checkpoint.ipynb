### Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- `fecha`:    10 lugares
- `producto`: 30 lugares
- `precio`:   10 lugares
- `cantidad`:  5 lugares

Hacer una funcion que cargue los datos en un DataFrame de Pandas.

```python
import pandas as pd

def cargar_datos():
    tamaño_columna = [10, 30, 10, 5]
    df= pd.read_fwf('datos.dat', widthas=tamaño_columna, names=['fecha', 'producto', 'precio', 'cantidad'])
    return df

datos = cargar_datos()
datos 
```

### Ejercicio 2: Calcular el total de ventas.
Hacer una función que sume los importes vendidos (precio * cantidad) y las cantidades.


```python
def calcular_totales(datos):
    datos['importe'] = datos['precio'] * datos['cantidad']
    totalidad_importe = datos['importe'].sum()
    totalidad_cantidad = datos ['cantidad'].sum()
    return totalidad_importe, totalidad_cantidad

importe, cantidad = calcular_totales(datos)

importe, cantidad = calcular_totales(datos)

print(f"Las ventas fueron de ${importe:2.2f} en {cantidad} unidades")
```

### Ejercicio 3: Listar las unidades vendidas.
Listar cuántas unidades se vendieron en total para cada producto


```python
def unidades_vendidas(datos):
    resultado = datos.groupby('producto')['cantidad'].sum().reset_index()
    resultado.colums = ['producto', 'Cantidad Vendida']
    print(resultado)

unidades_vendidas(datos)
```

###  Ejercicio 4: Listar el precio promedio por producto.
Hacer un listado del precio promedio por producto.


```python
def precio_promedio(datos):
    resultado = datos.groupby('producto')['precio'].mean().reset_index()
    resultado.columns = ['Producto', 'Precio Promedio']
    resultado['Precio Promedio'] = resultado['Precio Promedio'].round(2)
    print(resultado)

precio_promedio(datos)
```

### Ejercicio 5: Ranking de productos
Realizar un listado de los 3 productos más vendidos ordenados por la cantidad de unidades vendidas (ordenadas de mayor a menor)


```python
def ranking_productos(datos, top=3):
    resultado = datos.groupby('producto')['cantidad'].sum().reset_index()
    resultado = resultado.sort_values(by='cantidad', ascending=False).head(top)
    print(resultado)

ranking_productos(datos)
```

### Ejercicio 6: Listar las ventas por mes
Realizar un listado del total de unidades vendidas por producto separado por mes.


```python
def ventas_por_mes(datos):
    
    datos = datos[datos['fecha'].apply(lambda x: isinstance(x, str) and len(x) == 10)]

   
    datos['fecha'] = pd.to_datetime(datos['fecha'], format='%Y-%m-%d', errors='coerce')


    datos = datos.dropna(subset=['fecha'])

   
    datos['mes'] = datos['fecha'].dt.month

   
    resultado = datos.groupby(['producto', 'mes'])['cantidad'].sum().unstack(fill_value=0)

   
    resultado.columns = [f'Mes{col}' for col in resultado.columns]

   
    resultado = resultado.reset_index()

    print(resultado)


ventas_por_mes(datos)
```

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

```python
def resumen_ventas(datos):
    precio_promedio = datos.groupby('producto')['precio'].mean().reset_index()
    
    unidades_vendidas = datos.groupby('producto')['cantidad'].sum().reset_index()

    datos['importe'] = datos['precio'] * datos['cantidad']
    importe_total = datos.groupby('producto')['importe'].sum().reset_index()

    resumen = pd.merge(precio_promedio, unidades_vendidas, on= 'producto')
    resumen = pd.merge(resumen, importe_total, on ='producto')

    resumen.columns = ['Producto', 'Precio Promedio', 'Unidades vendidas', 'Importe Total']

    resumen = resumen.sort_values(by='Producto').reset_index(drop=True)

    print(resumen)
    

resumen_ventas(datos)
```

### Ejercicio 8

Escribe una función en Python que encuentre los valores de `a`, `b`, y `c` para que la función cuadrática `f(x) = a x^2 + b x + c` pase exactamente por los siguientes puntos:

| x  | y  |
|---:|---:|
|  0 |  0 |
|  1 |  8 |
|  2 | 12 |
|  3 | 12 |
|  5 |  0 |

### Requisitos:
- La función debe explorar posibles valores de `a`, `b`, y `c` utilizando un método de prueba y error.
- Debe devolver los valores que hagan que la diferencia entre la función `f(x)` y los valores medidos `y` sea exactamente cero para cada punto.

> **Pista**: Los valores de `a`, `b`, y `c` son números pequeños.

La idea es implementar el mismo algoritmo que se uso en el TP1 pero usando NumPy en lugar de Python puro.

```python
import numpy as np

def f(x, coeficientes):
    a, b, c = coeficientes
    return a * x**2 + b * x + c

def error(y, y_pred):
    return y - y_pred

def buscar_coeficientes():
    X = np.array([0, 1, 2, 3, 5])
    Y = np.array([0, 8, 12, 12, 0])

    a, b, c = 0, 0, 0
    coeficiente = np.array([a, b, c])

    for a in range(-10, 11):
        for b in range(-10, 11):  
            for c in range(-10, 11):
                coeficiente = np.array([a, b, c])
                y_pred = f(X, coeficiente)  
                if np.all(error(Y, y_pred) == 0):
                    return coeficiente  

    return None

coeficientes = buscar_coeficientes()
coeficientes
```

### Ejercicio 9: Resolver el ejercicio 3 del TP1 usando NumPy
Buscar los coeficientes de la función que minimice la suma de los cuadrados de las diferencias entre los valores medidos y los valores de la función.

1. Crear un array con los coeficientes elegidos al azar (usar `randint(-10,10,3)`).
2. Calcular el valor de la función y el error correspondiente.
3. Mientras que el error sea mayor a 1:
    1. Definir nuevos coeficientes agregándoles un pequeño valor al azar a los coeficientes actuales (aprendizaje = 0.001).
    2. Si el error para los nuevos coeficientes es menor que el anterior, reemplazar los coeficientes actuales por los nuevos.


```python
import numpy as np
from numpy.random import randint

def f(x, coeficientes):
    a,b,c = coeficientes
    return a*x**2 + b*x + c

def error(y, y_pred):
    return np.sum((y - y_pred)**2)

X = np.array([0, 1, 2, 3, 5])
Y = np.array([0, 8,12,11, 1]) # Observar que no son los mismos valores que en el ejemplo anterior

def buscar_coeficientes():
    coeficientes = np.random.randint(-10, 10, size=3)
    tasa_aprendisaje = 0.001
    error_actual = error(Y, f(X, coeficientes))

    while error_actual > 1:
        nuevos_coeficientes = coeficientes +np.random.randn(3) * tasa_aprendisaje

        nuevo_error = error(Y, f(X, nuevos_coeficientes))

        if nuevo_error < error_actual:
            coeficientes = nuevos_coeficientes
            error_actual = nuevo_error

    return coeficientes

coeficientes = buscar_coeficientes()
print(f"Los coeficientes son {coeficientes} y el error es {error(Y, f(X, coeficientes))}")


coeficientes = buscar_coeficientes()
coeficientes
```
