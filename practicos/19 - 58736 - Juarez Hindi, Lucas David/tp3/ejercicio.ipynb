## Ejercicio 1: Cargar Datos de ventas.

El archivo datos.dat contiene el registro de las ventas realizadas.

Tiene un formato de ancho fijo:
- fecha:    10 lugares
- producto: 30 lugares
- precio:   10 lugares
- cantidad:  5 lugares

hacer una funcion que carge los en una lista de diccionarios. 

nota: con 
```python
lineas = open("datos.dat", "r").readlines() 
```
puede obtener la lista de string de todos los datos 

````python
def cargar_ventas(archivo):
    ventas = []
    
    lineas = open(archivo, "r").readlines()
    
    for linea in lineas:
        fecha = linea[0:10].strip()
        producto = linea[10:40].strip()
        precio = float(linea[40:50].strip())
        cantidad = int(linea[50:55].strip())
        
      
        venta = {
            'fecha': fecha,
            'producto': producto,
            'precio': precio,
            'cantidad': cantidad
        }
        
        ventas.append(venta)
    
    return ventas


ventas = cargar_ventas("datos.dat")
print(ventas)


````

## Ejercicio 2: Calcular el total de ventas.

Hacer una funcion que sume los importes vendidos (precio * cantidad) y las cantidades

````python
def calcular_totales(datos):
    total_importe = 0
    total_cantidad = 0
    
    for venta in datos:
        precio = venta['precio']
        cantidad = venta['cantidad']
        total_importe += precio * cantidad
        total_cantidad += cantidad
        
    return total_importe, total_cantidad

def leer_datos(archivo):
    datos = []
    with open(archivo, 'r') as file:
        for line in file:
            line = line.strip()
            if line:  
                partes = line.split()
                fecha = partes[0]  # La fecha es la primera parte
                producto = " ".join(partes[1:-2])  
                precio = int(partes[-2])  
                cantidad = int(partes[-1]) 
                
                
                datos.append({
                    'fecha': fecha,
                    'producto': producto,
                    'precio': precio,
                    'cantidad': cantidad
                })
    return datos


datos = leer_datos('datos.dat')


importe, cantidad = calcular_totales(datos)


print(f"Las ventas fueron de ${importe:,.2f} en {cantidad} unidades")

````

## Ejercicio 3: Listar las unidades vendidas.

Listar cuantas unidades se vendieron en total para cada producto

````python
def unidades_vendidas(archivo):
    ventas = {}
    with open(archivo, 'r') as f:
        for linea in f:
            partes = linea.strip().split()  
            producto = partes[1]  
            cantidad = int(partes[-1])  
            
            if producto in ventas:
                ventas[producto] += cantidad
            else:
                ventas[producto] = cantidad
    return ventas

def listar_ventas(ventas):
    for producto, cantidad in ventas.items():  
        print(f"Producto: {producto}, Unidades vendidas: {cantidad}")

archivo = 'datos.dat'

ventas = unidades_vendidas(archivo)

listar_ventas(ventas)

````

##  Ejercicio 4: Listar el precio promedio por producto.

Hacer un listado del precio promedio por productos.

````python
def leer_datos(archivo):
    datos = []
    with open(archivo, 'r') as f:
        for linea in f:
            linea = linea.strip()  
            if not linea: 
                continue
            
            if len(linea) < 50:
                print(f"Línea malformada ignorada: {linea}")
                continue
            
            fecha = linea[:10].strip()  
            nombre_producto = linea[10:40].strip() 
            try:
                precio = int(linea[40:45].strip())  
                cantidad = int(linea[45:50].strip())  
            except ValueError:
                print(f"Error al convertir precios/cantidades: {linea}")
                continue
            
            datos.append((nombre_producto, precio, cantidad))
    return datos

def precio_promedio(datos):
    suma_precios = {}
    conteo_precios = {}

    
    for producto, precio, cantidad in datos:
        if producto in suma_precios:
            suma_precios[producto] += precio * cantidad  
            conteo_precios[producto] += cantidad  
        else:
            suma_precios[producto] = precio * cantidad
            conteo_precios[producto] = cantidad

    precios_promedio = {producto: suma / conteo_precios[producto] for producto, suma in suma_precios.items()}
    
    return precios_promedio

def listar_precios(precios):
    print("Precio promedio por producto:")
    for producto, precio in precios.items():
        print(f"{producto}: {precio:.2f}")

datos = leer_datos('datos.dat')
precios = precio_promedio(datos)
listar_precios(precios)
````

### Ejercicio 5: Ranking de productos

Realizar un listado de los 3 productos mas vendidos ordenados la cantidad de unidades vendidas (ordenadas de mayor a menor)


````python
def leer_datos(archivo):
    datos = []
    with open(archivo, 'r') as file:
        for linea in file:
            partes = linea.strip().split()
            nombre_producto = ' '.join(partes[:-2])
            unidades_vendidas = int(partes[-1])
            datos.append({'nombre': nombre_producto, 'unidades': unidades_vendidas})
    return datos

def ranking_productos(datos, top=3):
    ventas = {}
    for producto in datos:
        nombre = producto['nombre']
        unidades = producto['unidades']
        if nombre in ventas:
            ventas[nombre] += unidades
        else:
            ventas[nombre] = unidades

    ranking = sorted(ventas.items(), key=lambda x: x[1], reverse=True)
    return ranking[:top]

def listar_ranking(ranking):
    for i, (producto, unidades) in enumerate(ranking, start=1):
        print(f"{i}. {producto}: {unidades} unidades vendidas")

datos = leer_datos('datos.dat')

ranking = ranking_productos(datos)
listar_ranking(ranking)
````

### Ejercicio 6: Listar las ventas por mes

Realizar un listado del total de unidades vendidas por productos separada por mes.

````python
from collections import defaultdict
from datetime import datetime

def leer_datos(archivo):
    datos = []
    with open(archivo, 'r') as file:
        for linea in file:
            fecha = linea[:10]  
            resto = linea[10:]  

            partes = resto.split()
            nombre_producto = ' '.join(partes[:-1])  
            unidades_vendidas = int(partes[-1])  

            datos.append({'fecha': fecha, 'nombre': nombre_producto, 'unidades': unidades_vendidas})
    return datos

def ventas_por_mes(datos):
    ventas = defaultdict(lambda: defaultdict(int))

    for producto in datos:
        nombre = producto['nombre']
        unidades = producto['unidades']
        fecha = producto['fecha']
        mes_anio = datetime.strptime(fecha, '%Y-%m-%d').strftime('%Y-%m')

        ventas[mes_anio][nombre] += unidades

    return dict(ventas)

def listar_ventas_mensuales(ventas):
    for mes, productos in sorted(ventas.items()):
        print(f"Ventas para el mes: {mes}")
        for nombre, unidades in productos.items():
            print(f"  {nombre}: {unidades} unidades vendidas")
        print()  # Línea en blanco entre meses

datos = leer_datos('datos.dat')

ventas = ventas_por_mes(datos)
listar_ventas_mensuales(ventas)
````

### Ejercicio 7: Informe general

Mostrar un listado de productos ordenados alfabeticamente que contengan el precio promedio, la cantidad de unidades vendidas y el importe total vendido para cada producto

````python
import re

def leer_datos(archivo):
    datos = []
    with open(archivo, 'r') as f:
        for linea in f:
            linea = linea.strip()  
            match = re.match(r'(\d{4}-\d{2}-\d{2})(\S+(?:\s+\S+)*)\s+(\d+)\s+(\d+)', linea)
            if match:
                fecha = match.group(1)
                producto = match.group(2)
                precio = int(match.group(3))
                cantidad = int(match.group(4))
                
                datos.append({'fecha': fecha, 'producto': producto, 'precio': precio, 'cantidad': cantidad})
            else:
                print(f"Formato incorrecto en la línea: {linea}")
    
    print(f"Datos leídos: {datos}")  
    return datos

def resumen_ventas(datos):
    resumen = {}
    
    for registro in datos:
        producto = registro['producto']
        precio = registro['precio']
        cantidad = registro['cantidad']
        
        if producto not in resumen:
            resumen[producto] = {'total_vendido': 0, 'cantidad_vendida': 0}
        
        resumen[producto]['total_vendido'] += precio * cantidad
        resumen[producto]['cantidad_vendida'] += cantidad
    
    for producto, info in resumen.items():
        info['precio_promedio'] = info['total_vendido'] / info['cantidad_vendida']
    
    print(f"Resumen de ventas: {resumen}")  
    return resumen

def informe_ventas(resumen):
    productos_ordenados = sorted(resumen.keys())
    
    print("Listado de Productos:")
    print(f"{'Producto':<30} {'Precio Promedio':<20} {'Cantidad Vendida':<20} {'Importe Total':<20}")
    
    for producto in productos_ordenados:
        info = resumen[producto]
        print(f"{producto:<30} {info['precio_promedio']:<20.2f} {info['cantidad_vendida']:<20} {info['total_vendido']:<20.2f}")

datos = leer_datos('datos.dat')

resumen = resumen_ventas(datos)
informe_ventas(resumen)
````
